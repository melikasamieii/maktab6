```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

data = load_breast_cancer()

df = pd.DataFrame(data.data , columns=data.feature_names)
df["label"]= data.target_names[data.target]
df.head(20)


```

```python
ben = df[df["label"]== "benign"].head(3)
mal = df[df["label"]== "malignant"].head(3)
samples= pd.concat([ben, mal])
X = samples.drop(columns= ["label"]).values
print(X.shape)
samples.describe()
```

```python
from scipy.spatial.distance import euclidean, cityblock, chebyshev, cosine, mahalanobis
import numpy as np

X_full = data.data
cov = np.cov(X_full.T)
# transpose(30,6), mahalanobis, if X dis be var--> singular matrix value-(inv bshe pinv)
inv_cov = np.linalg.inv(cov)
# mahalanobis, inverse  

rows = []

for i in range(len(X)):
    for j in range(i+1, len(X)):
        rows.append({
    "pair": f"{i}-{j}",
    "label_i": samples.iloc[i]["label"],
    "label_j": samples.iloc[j]["label"],
    "Manhattan": cityblock(X[i], X[j]),
    "Euclidean": euclidean(X[i], X[j]),
    "Chebyshev": chebyshev(X[i], X[j]),
    "Cosine": cosine(X[i], X[j]),
    "Mahalanobis": mahalanobis(X[i], X[j], inv_cov)
    })

df_dist = pd.DataFrame(rows)
df_dist
    
```

فاصله با انتخاب نوع محاسبه تغییر میکنه پس ثابت نیست
مقیاس های بزرگ مثل area روی فاصله هایی مثل اقلیدسی اثر دارن و ممکنه فاصله درست نتونیم بدست بیاریم
برای euclidean, mahalanobis  مثلا نسبت ویزگی ها/ زاویه با توجه به مقیاس و یا پراکندگی مهمه
اگر فاصله اشتباه انتخاب شه لیبل گذاری اشتباه و جان فردی در خطر

در جدول هر جا:

فاصله کوچک باشد

 دو نمونه شبیه هستند (اغلب داخل یک کلاس)

فاصله بزرگ باشد

 دو نمونه متفاوت هستند (اغلب بین دو کلاس)

Euclidean — فاصلهٔ اقلیدسی
 فقط اختلاف عددی را می‌سنجد، آن هم با مربع
  ویژگی‌های بزرگ مثل area, radius خراب میکند
به همین دلیل فاصلهٔ بین دو کلاس در جدول خیلی بزرگ است و این متریک تفاوت کلاس‌ها را خوب نشان می‌دهد
اختلاف بین کلاس‌ها شدید است، پس  قدرت تفکیک خوبی دارد اما بیش‌ازحد به scale وابسته است.


 مزایا:

اختلاف واقعی روی ویژگی‌ها را خوب می‌گیرد

بین benign و malignant معمولاً فاصلهٔ زیادی می‌اندازد → قدرت تفکیک بالا

 ریسک:

به شدت به مقیاس ویژگی‌ها وابسته است
فاصله را غیرواقعی بزرگ می‌کنند

Manhattan — فاصلهٔ L1
اصله‌های بین کلاس‌ها چندین برابر داخل‌کلاسی هستند

فاصله‌های داخل benign کوچک‌ترین مقادیر جدول هستند

ون Manhattan هم scale را در نظر می‌گیرد، ساختار مشابه Euclidean دارد.
فقط به‌جای مربع اختلاف‌ها را مستقیم جمع می‌کند  بنابراین کمی ملایم‌تر است.
فاصلهٔ مقیاس‌محور


مزایا:

اختلاف‌ها را ملموس‌تر از Euclidean می‌سنجد

به outlier کمتر حساس است


 ریسک:

همچنان مقیاس ویژگی‌ها فاصله را کنترل می‌کند

همان مشکل Euclidean را دارد → خطر misclassification

Chebyshev — بیشترین اختلاف روی یک ویژگی

اصلهٔ benign–benign خیلی کم است (چون همه ویژگی‌ها نزدیکند)
فاصلهٔ malignant–malignant هم کم است

اما فاصلهٔ benign–malignant خیلی بزرگ می‌شود اگر فقط یک feature تفاوت بزرگ داشته باشد

این فاصله فقط به «بدترین اختلاف» نگاه می‌کند، و در داده‌هایی که scale متفاوت دارند، بین کلاس‌ها شکاف خیلی بزرگ ایجاد می‌کند.



مزایا:

اختلاف‌های «خیلی بزرگ» را خوب تشخیص می‌دهد


 ریسک خیلی بزرگ:

فقط بزرگ‌ترین اختلاف روی یک ویژگی را نگاه می‌کند


Cosine distance — بررسی زاویه (جهت)

فاصله‌ها بسیار کوچک‌اند (مثلاً چیزی بین ۰ و ۰.۰۳)

فاصلهٔ داخل benign تقریباً برابر فاصلهٔ benign–malignant است
 یعنی تفکیک کلاس‌ها ضعیف است

  Cosine برای این دیتا مناسب نیست؛ چون اندازه توده مهم است ولی آن را نادیده می‌گیرد.


  ریسک کشنده:

اندازهٔ واقعی توده را نمی‌بیند
در Breast Cancer تشخیص malignant همین اندازه است

دو نمونه‌ی کاملاً متفاوت از نظر size
اگر نسبت ویژگی‌هاشان مشابه باشد  cosine می‌گوید «خیلی شبیه‌اند»

 Mahalanobis — بهترین توزیع واقعی فاصله

 
فاصلهٔ benign–benign کوچک

فاصلهٔ malignant–malignant کمی بزرگ‌تر

فاصلهٔ benign–malignant واضحاً از داخل‌کلاس‌ها بزرگ‌تر
اما نه به اندازه‌ی Euclidean که کنترلش از دست بره.

فاصلهٔ واقعی را بر اساس پراکندگی (variance) و همبستگی(features) حساب می‌کند.
یعنی اگر یک ویژگی ذاتاً بزرگ باشد (area)، وزنش کمتر می‌شود.


سناریوی «سیستم هشدار اولیه بدخیمی»

و انتخاب بهترین فاصله + مزایا + ریسک‌ها


 فاصلهٔ این بیمار با نمونه‌های قبلی را نگاه می‌کنیم
اگر به malignant نزدیک‌تر باشد مشکوک به بدخیمی

scale واقعی ویژگی‌ها را در نظر بگیرد

ویژگی‌های بزرگ را کنترل کند

همبستگی بین ویژگی‌ها را بشناسد

جهت‌های طبیعی تغییر داده را بداند

و فاصله را بی‌خود بزرگ یا کوچک نکند

یک benign کوچک و یک malignant کوچک را مثل هم ببیند

یا یک benign بزرگ را اشتباهی مشکوک کند

Euclidean اگر بدون نرمال‌سازی استفاده شود، غیرقابل اعتماد و خطرناک است.
